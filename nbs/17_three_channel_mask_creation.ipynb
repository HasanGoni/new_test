```python
#| export
def load_model(config):
    model = tf.keras.models.load_model(config['model']['save_path'] )
    return model
```

```python
#| export
def predct_msk(
        model:tf.keras.Model,
        im_file:str,
        threshold:float=0.5,
        save_path:str=None
  ):

  img = read_normalize(im_file=im_file, config=config)
  pred = model.predict(img)
  msk = pred[0] > threshold
  msk = msk * 255.0
  msk = msk.astype(np.uint8)
  if save_path:
    cv2.imwrite(save_path, msk)
```

```python
model = tf.keras.models.load_model(
    f"{config['model']['save_path']}/{config['model']['model_name']}"
    ,compile=False
)
```

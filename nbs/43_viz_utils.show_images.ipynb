```python
#| export
import sys
from pathlib import Path
import matplotlib.pyplot as plt
import numpy as np
```

```python
#| export
import albumentations as A
from typing import Dict, Any, Tuple

class ShadedMaskAugmentation(A.ImageOnlyTransform):
    """Custom Albumentations transform that adds shading only to masked regions of grayscale images"""
    
    def __init__(
        self,
        alpha: float = 0.3,
        intensity: float = 255.0, # Intensity of shading for grayscale
        always_apply: bool = False,
        p: float = 0.5
    ):
        super().__init__(always_apply, p)
        self.alpha = alpha
        self.intensity = intensity
        
    def apply(self, image: np.ndarray, mask: np.ndarray = None, **params) -> np.ndarray:
        """Apply shading to masked regions
        Args:
            image: Input grayscale image array (H,W)
            mask: Binary mask array (H,W) or (N,H,W)
        Returns:
            Augmented grayscale image with shading in masked regions
        """
        # Convert image to float32 [0,1] range
        image = image.astype(np.float32) / 255.0
        
        # Handle both single and multiple masks
        if len(mask.shape) == 2:
            mask = mask[np.newaxis,...]
            
        # Create shaded overlay for each mask
        overlay = np.zeros_like(image)
        for m in mask:
            overlay = np.where(m, self.intensity/255.0, overlay)
            
        # Blend original image with shaded overlay
        augmented = image * (1-self.alpha) + overlay * self.alpha
        
        # Clip values and convert back to uint8
        augmented = np.clip(augmented * 255, 0, 255).astype(np.uint8)
        
        return augmented

    def get_transform_init_args_names(self) -> Tuple[str, ...]:
        """Get transform init args names"""
        return ("alpha", "intensity")

# Example training pipeline with the custom augmentation
def get_training_augmentation(
    height: int = 512,
    width: int = 512
) -> A.Compose:
    """Get training augmentation pipeline including shaded mask augmentation
    
    Args:
        height: Target height
        width: Target width
        
    Returns:
        Albumentations Compose object with augmentation pipeline
    """
    train_transform = [
        A.RandomResizedCrop(height=height, width=width, scale=(0.8, 1.0), p=1.0),
        A.HorizontalFlip(p=0.5),
        A.VerticalFlip(p=0.5),
        ShadedMaskAugmentation(p=0.5),
        A.Normalize(mean=[0.485], std=[0.229]), # Single channel normalization
    ]
    return A.Compose(train_transform)

def get_valid_augmentation(
    height: int = 512,
    width: int = 512
) -> A.Compose:
    """Get validation augmentation pipeline
    
    Args:
        height: Target height
        width: Target width
        
    Returns:
        Albumentations Compose object with validation pipeline
    """
    test_transform = [
        A.Resize(height=height, width=width),
        A.Normalize(mean=[0.485], std=[0.229]), # Single channel normalization
    ]
    return A.Compose(test_transform)
```

```python
#| export
def show_three_images(
    im1:np.ndarray,  # First image array
    im2:np.ndarray,  # Second image array
    im3:np.ndarray,  # Third image array
    title1:str='Image 1',  # Title for first image
    title2:str='Image 2',  # Title for second image
    title3:str='Image 3',  # Title for third image
    figsize:tuple=(20,7),  # Larger figure size for better clarity
    layout:str='row',  # Layout type: 'row' or 'column'
    scale:float=1.0  # Scale factor to resize images
):
    'Show three images in a single row or column with titles, optimized for clarity'
    plt.figure(figsize=figsize, dpi=150)  # Higher DPI for better resolution
    
    # Scale images if needed
    if scale != 1.0:
        im1 = im1 * scale
        im2 = im2 * scale
        im3 = im3 * scale
    
    if layout == 'row':
        plt.subplot(131)
        plt.imshow(im1)
        plt.title(title1, pad=10, fontsize=12)
        plt.axis('off')
        
        plt.subplot(132)
        plt.imshow(im2)
        plt.title(title2, pad=10, fontsize=12)
        plt.axis('off')
        
        plt.subplot(133)
        plt.imshow(im3)
        plt.title(title3, pad=10, fontsize=12)
        plt.axis('off')
    else:  # column layout
        plt.subplot(311)
        plt.imshow(im1)
        plt.title(title1, pad=10, fontsize=12)
        plt.axis('off')
        
        plt.subplot(312)
        plt.imshow(im2)
        plt.title(title2, pad=10, fontsize=12)
        plt.axis('off')
        
        plt.subplot(313)
        plt.imshow(im3)
        plt.title(title3, pad=10, fontsize=12)
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
```

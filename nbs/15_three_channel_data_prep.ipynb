```python
def cut_image(img):
    height, width = img.shape[:2]
    half_height, half_width = height // 2, width // 2

    top_left = img[:half_height, :half_width]
    top_right = img[:half_height, half_width:]
    bottom_left = img[half_height:, :half_width]
    bottom_right = img[half_height:, half_width:]
    return top_left, top_right, bottom_left, bottom_right
```

```python
#| export
def recipe_wise_sorting(path:str):
    'group files based on recipe'
    recipes = []
    recipe_keys={}
    recipe_groups = ['108', '109', '122', '17', '18', '50', '51', '57', '68', '87']
    for j in recipe_groups:
        recipe_list  = []
        for i in im_src.ls():
            recipe = get_m_name(f'{i}')


            if recipe == j:
                recipe_list.append(i)
        recipe_keys[j] = recipe_list
    return recipe_keys
    
```

```python
recipe_keys = recipe_wise_sorting(im_src)
for i in tqdm(recipe_keys.keys(), total=len(recipe_keys.keys())):
    fn_ = np.random.choice(recipe_keys[i], 1)[0]
    im_name = Path(fn_).name
    #msk_name =f'{msk_src}/{im_name}'
    shutil.copyfile(f'{im_src}/{im_name}', f'{im_dst}/{im_name}')
    shutil.copyfile(f'{msk_src}/{im_name}', f'{msk_dst}/{im_name}')
   

       
        
```

```python
pn_cnt_dict_ = pin_count_dict()
fns_ = []
for i in tqdm(msk_src.ls(),total=len(msk_src.ls())):
    recipe_name = get_m_name(f'{i}')

    if recipe_name != '9999999':
        cnt = pn_cnt_dict_[recipe_name]
        expected = cnt
        real_ = pin_count_from_fn(i)
    
        
        if int(expected) != int(real_):
            print(expected, real_)
            fns_.append(i)
            
```

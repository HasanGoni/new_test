# Train HPC

> Hydra entry point for training the 3-channel binary pin segmentation model on HPC.
>
> After `nbdev_export`, run:
> ```bash
> python end_test_pin_detection_patch/scripts/train_hpc.py \
>     root=/path/to/patch_dataset \
>     checkpoint_dir=checkpoints/run1 \
>     num_epochs=60 batch_size=16
> ```

````python
#| export
import sys
from pathlib import Path

import hydra
from omegaconf import DictConfig, OmegaConf
````

````python
#| export
from end_test_pin_detection_patch.preprocessing.three_channel_image_train import (
    get_model,
    train,
    set_seed,
)
````

````python
#| export
@hydra.main(
    version_base=None,
    config_path=str(Path(__file__).resolve().parents[2] / "configs"),
    config_name="train_hpc",
)
def main(cfg: DictConfig) -> None:
    "Hydra entry point: build model from cfg and run training loop"
    print("=" * 70)
    print("TRAIN HPC - CONFIGURATION")
    print("=" * 70)
    print(OmegaConf.to_yaml(cfg))
    print("=" * 70 + "\n")

    set_seed(42)

    model = get_model(
        arch=cfg.arch,
        pretrained=cfg.pretrained,
        in_channels=cfg.in_channels,
    )

    train(
        root=cfg.root,
        model=model,
        num_epochs=cfg.num_epochs,
        batch_size=cfg.batch_size,
        encoder_lr=cfg.encoder_lr,
        decoder_lr=cfg.decoder_lr,
        freeze_epochs=cfg.freeze_epochs,
        crop_size=cfg.crop_size,
        device_str=cfg.device_str,
        checkpoint_dir=cfg.checkpoint_dir,
        use_amp=cfg.use_amp,
        save_every=cfg.save_every,
        resume=cfg.resume,
    )
````

````python
#| export
if __name__ == "__main__":
    main()
````
